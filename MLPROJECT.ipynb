```python
import ast
# CREATING SIMPLE LIST FOR GENRES
def convert(obj):
  L=[]
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
import ast
# GERNERATING THE NAMES F TOP 3 ACTORS(ORIGINAL NAME)
def convert3(obj):
  L=[]
  count=0
  for i in ast.literal_eval(obj):
    if count!=3:
      L.append(i['name'])
      count+=1
    else:
      break
  return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
#  GETTING ALL THE DIRECTORS
def get_director(obj):
  L=[]
  for i in ast.literal_eval(obj):
    if i['job']=='Director':
      L.append(i['name'])
      break
  return L
```

```python
movies['crew']=movies['crew'].apply(get_director)
```

```python
movies['crew'].head()
```

```python
# Concating the overview
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
# removing the spaces between the names

movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
# ADDING TAGS
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
import nltk
```

```python
#@title Default title text
# STEMMING USING NLTK

from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
# Stemming function
def dostemming(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
#GETTING COSINE DISTANCE

from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
# sorted(similarity[7],reverse=True)
```

```python
# GETTING SIMILAR MOVIES

def recomend(movie):
  movie_index=newdata[newdata['title']==movie].index[0]
  distances=similarity[movie_index]
  movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:10]
  for i in movies_list:
    print(newdata.iloc[i[0]].title)
  return 
```
